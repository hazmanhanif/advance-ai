# **Extract features (sepal length, sepal width, petal length, petal width)**


```python
X = iris_data.iloc[:, :-1].values
print(X)
```

# **Explanation**


* iris_data is assumed to be a Pandas DataFrame containing the Iris dataset, and it's expected that the last column represents the class labels.

*   iloc is a method in Pandas used for integer-location based indexing. In this case, it's used to select specific rows and columns from the DataFrame.

*   [:, :-1] specifies that we want to select all rows (:) and all columns except the last one ([:-1]).

*   .values is used to convert the selected data into a NumPy array.



# **Standardize**

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# StandardScaler is a class from scikit-learn's preprocessing module. It is used for standardizing features by removing the mean and scaling to unit variance.
# scaler is an instance of the StandardScaler class.

# The fit_transform method of the StandardScaler class is used to standardize the feature matrix X.
# fit_transform performs two steps:
# It computes the mean and standard deviation of each feature in X using the fit method.
# It then applies the transformation by subtracting the mean and dividing by the standard deviation for each feature using the transform method.
# The result is a new standardized feature matrix X_scaled. Each feature in X_scaled has a mean of 0 and a standard deviation of 1.
```

# **Explanation**
* StandardScaler is a class from scikit-learn's preprocessing module. It is used for standardizing features by removing the mean and scaling to unit variance.
* scaler is an instance of the StandardScaler class.

* The fit_transform method of the StandardScaler class is used to standardize the feature matrix X.
* fit_transform performs two steps:
* It computes the mean and standard deviation of each feature in X using the fit method.
* It then applies the transformation by subtracting the mean and dividing by the standard deviation for each feature using the transform method.
* The result is a new standardized feature matrix X_scaled. Each feature in X_scaled has a mean of 0 and a standard deviation of 1.

# **6 Different Algorithm**

```python
# Apply K-Means clustering
kmeans = KMeans(n_clusters=3, random_state=42)
iris_data['kmeans_cluster'] = kmeans.fit_predict(X_scaled)
# Applying K-Means clustering with 3 clusters to the standardized features.


# Apply DBSCAN clustering
dbscan = DBSCAN(eps=0.5, min_samples=5)
iris_data['dbscan_cluster'] = dbscan.fit_predict(X_scaled)
# Applying DBSCAN clustering to the standardized features.

# Apply Birch clustering
birch = Birch(n_clusters=3)
iris_data['birch_cluster'] = birch.fit_predict(X_scaled)
# Applying Birch clustering with 3 clusters to the standardized features.

# Apply Gaussian Mixture Model (GMM) clustering
gmm = GaussianMixture(n_components=3)
iris_data['gmm_cluster'] = gmm.fit_predict(X_scaled)
# Applying GMM clustering with 3 components to the standardized features.

# Apply Agglomerative Clustering
agglomerative = AgglomerativeClustering(n_clusters=3)
iris_data['agglomerative_cluster'] = agglomerative.fit_predict(X_scaled)
# Applying Agglomerative Clustering with 3 clusters to the standardized features.

# Apply Spectral Clustering
spectral = SpectralClustering(n_clusters=3, affinity='nearest_neighbors', random_state=100)
iris_data['spectral_cluster'] = spectral.fit_predict(X_scaled)
# Applying Spectral Clustering with 3 clusters to the standardized features.

```

# **K-Means Clustering**
* K-Means: Initializes the KMeans clustering algorithm with n_clusters=3, meaning it will attempt to partition the data into 3 clusters.
* fit_predict: Fits the KMeans model to the standardized feature matrix X_scaled and assigns each data point to a cluster. The cluster labels are then added as a new column 'kmeans_cluster' to the iris_data DataFrame.

# **DBSCAN Clustering**
* DBSCAN: Initializes the DBSCAN clustering algorithm with eps=0.5 (maximum distance between two samples for one to be considered as in the neighborhood of the other) and min_samples=5 (the number of samples in a neighborhood for a point to be considered as a core point).
* fit_predict: Fits the DBSCAN model to the standardized feature matrix X_scaled and assigns each data point to a cluster. The cluster labels (including noise points labeled as -1) are added as a new column 'dbscan_cluster' to the iris_data DataFrame.

# **Birch Clustering**
* Birch: Initializes the Birch clustering algorithm with n_clusters=3, indicating it will attempt to partition the data into 3 clusters.
* fit_predict: Fits the Birch model to the standardized feature matrix X_scaled and assigns each data point to a cluster. The cluster labels are then added as a new column 'birch_cluster' to the iris_data DataFrame.

# **Gaussian Mixture Clustering**
* Gaussian Mixture: Initializes the GMM clustering algorithm with n_components=3, specifying the number of Gaussian components (clusters) in the mixture.
* fit_predict: Fits the GMM model to the standardized feature matrix X_scaled and assigns each data point to a cluster. The cluster labels are then added as a new column 'gmm_cluster' to the iris_data DataFrame.

#**Agglomerative Clustering**
* Agglomerative Clustering: Initializes the Agglomerative Clustering algorithm with n_clusters=3, indicating it will form 3 clusters.
* fit_predict: Fits the Agglomerative Clustering model to the standardized feature matrix X_scaled and assigns each data point to a cluster. The cluster labels are then added as a new column 'agglomerative_cluster' to the iris_data DataFrame.

# **Spectral Clustering**
* Spectral Clustering: Initializes the Spectral Clustering algorithm with n_clusters=3, using nearest neighbors for affinity (similarity measure between samples).
* fit_predict: Fits the Spectral Clustering model to the standardized feature matrix X_scaled and assigns each data point to a cluster. The cluster labels are then added as a new column 'spectral_cluster' to the iris_data DataFrame.

# **Visualize**

```python
# Visualize the clusters using PCA for dimensionality reduction
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
iris_data['pca1'] = X_pca[:, 0]
iris_data['pca2'] = X_pca[:, 1]
```

# **Explanation**

* PCA: Principal Component Analysis is a technique for reducing the dimensionality of the dataset while retaining as much variance as possible. It transforms the original features into a new set of uncorrelated features, called principal components.

* n_components=2: In this context, PCA is configured to reduce the dimensionality to 2. It means that the transformed dataset will have two principal components.

* The fit_transform method of the PCA object is applied to the standardized feature matrix X_scaled. This method computes the principal components and transforms the data into the new space defined by these components.

* The result is stored in the variable X_pca, which now contains the dataset in the reduced-dimensional space.

* X_pca[:, 0] extracts the values of the first principal component for each data point and assigns them to the 'pca1' column.

* X_pca[:, 1] extracts the values of the second principal component for each data point and assigns them to the 'pca2' column.

# **Constant**

```python
# Constants for Visualization
def visualize_clusters_2d(X, labels, xlabel, ylabel, title):
    plt.scatter(X[xlabel], X[ylabel], c=labels, cmap='viridis', edgecolor='k')
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.show()
```

# **Explanation**

* plt.scatter(): This function is used to create a scatter plot.
* X[xlabel] and X[ylabel]: Extracts the values of the specified columns from the input data X. These columns are used as the x and y coordinates for each data point in the scatter plot.
* c=labels: The color of each point is determined by the cluster labels. Different clusters are assigned different colors.
* cmap='viridis': Specifies the color map to be used. 'viridis' is a perceptually uniform color map.
* edgecolor='k': Sets the edge color of the points to black.

* plt.title(): Sets the title of the plot to the specified title.
* plt.xlabel(): Sets the label for the x-axis to the specified xlabel.
* plt.ylabel(): Sets the label for the y-axis to the specified ylabel.

* plt.show(): Displays the generated scatter plot.

# **Result**

```python
# Result Visualization for each clustering algorithm
visualize_clusters_2d(iris_data, iris_data['kmeans_cluster'], 'pca1', 'pca2', 'K-Means Clustering of Iris Dataset')
visualize_clusters_2d(iris_data, iris_data['dbscan_cluster'], 'pca1', 'pca2', 'DBSCAN Clustering of Iris Dataset')
visualize_clusters_2d(iris_data, iris_data['birch_cluster'], 'pca1', 'pca2', 'Birch Clustering of Iris Dataset')
visualize_clusters_2d(iris_data, iris_data['gmm_cluster'], 'pca1', 'pca2', 'GMM Clustering of Iris Dataset')
visualize_clusters_2d(iris_data, iris_data['agglomerative_cluster'], 'pca1', 'pca2', 'Agglomerative Clustering of Iris Dataset')
visualize_clusters_2d(iris_data, iris_data['spectral_cluster'], 'pca1', 'pca2', 'Spectral Clustering of Iris Dataset')

```
